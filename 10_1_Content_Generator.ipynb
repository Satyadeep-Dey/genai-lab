```python
# Constants
GPT_41_mini = "gpt-4.1-mini"
GPT_52 = "gpt-5.2"
```

```python
# Sign in to HuggingFace Hub

hf_token = userdata.get('HF_TOKEN')
login(hf_token, add_to_git_credential=True)
```

```python
# Sign in to OpenAI using Secrets in Colab

openai_api_key = userdata.get('OPENAI_API_KEY')
openai = OpenAI(api_key=openai_api_key)
```

```python
book_name = "A Tale of Two Cities "
```

```python
system_message = """You are an assistant who is given the name of a novel or short story and you summarize it and return data in markdown format.
Your summary should be elaborate so that the story can be understood and should include details of characters, places and incidents.
Your summary should include the following :
  1. Name of the novel or short story.
  2. Name of the author.
  3. An overview of the story .
  4. Major characters and their detail .
  5. Include all the chapters.Each chapter should be summarized with 4 bullet points as shown in below example :
    Here is the example ->
    Chapter 3: The Shadow
    • Madame Defarge visits Lucie coldly, under the pretense of concern.
    • Lucie senses danger—Madame is “the shadow”, symbolizing coming death.
    • Carton and Jarvis Lorry also arrive in Paris, determined to help.
    • Madame Defarge's motives are becoming clear: her interest in Lucie is not sympathy—it's surveillance.
  """

user_prompt = "Here is the name of the novel or short story  : " + book_name

messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]

display(messages)
```

```python
completion = openai.chat.completions.create(
        model=GPT_41_mini,
        messages=messages,
        temperature = 0.1
    )
```

```python
summary = completion.choices[0].message.content
display(summary)
```

```python
write_text_to_file("Files/Knowledge-Base", "OpenAI _generated by gpt-4.1-mini _TOTC.txt", summary)
```

```python
completion = openai.chat.completions.create(
        model=GPT_52,
        messages=messages,
        temperature = 0.1
    )
```

```python
summary = completion.choices[0].message.content
display(summary)
```
