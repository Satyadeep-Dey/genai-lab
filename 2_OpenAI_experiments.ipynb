```python
# Import required libraries

import requests
import json
from openai import OpenAI
from google.colab import userdata

#define constants
#MODEL = "gpt-4o-mini"
MODEL = "gpt-4.1-mini"
#MODEL = "gpt-5.2"
```

```python
#This is how we programatically call ChatGPT

def message_gpt(system_message,prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    completion = openai.chat.completions.create(
        model=MODEL,
        messages=messages,
    )
    return completion.choices[0].message.content
```

```python
system_message = "You are a helpful assistant"
prompt = "what is the capital of Germany ? "
message_gpt(system_message,prompt)
```

```python
#Genarate Image using Open AI

from openai import OpenAI
import base64
from PIL import Image
import io

client = OpenAI(api_key=openai_api_key)

prompt = "A futuristic AI engineer working on stock prediction models, cinematic lighting, ultra detailed"

result = client.images.generate(
    model="gpt-image-1",
    prompt=prompt,
    size="1024x1024"
    #background="transparent"
)

# The image is returned as base64
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)

# Display in Colab
image = Image.open(io.BytesIO(image_bytes))
display(image)
```

```python
with open("generated_image.png", "wb") as f:
    f.write(image_bytes)
```

```python
# Text → Speech using OpenAI
import os
from IPython.display import Audio

client = OpenAI(api_key=openai_api_key)

speech_file_path = "speech.mp3"

response = client.audio.speech.create(
    model="gpt-4o-mini-tts",
    #voice="alloy",   # you can try other voices
    voice="ballad",
    input="My name is Sachin Tendulkar. I am a cricketer and I played for Mumbai and India."
)

# Write to file as you have to save it before playing
with open(speech_file_path, "wb") as f:
    f.write(response.content)

#Play Audio in Colab
Audio(speech_file_path)
```
