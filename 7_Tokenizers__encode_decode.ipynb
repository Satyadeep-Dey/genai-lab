# Tokenizers

In this notebook , we use the transformers tokenizer to encode and decode using 3 different models .

Tokenizer performs two related operations:

*   Encode: Text → Token IDs
*   Decode: Token IDs → Text

***Note : When we call a LLM e.g., ChatGPT , simplified pipeline looks like this***

```
Your Question (text)
        ↓
Tokenizer.encode()
        ↓
Token IDs (integers)
        ↓
Embedding Layer
        ↓
Vectors
        ↓
Transformer (LLM)
        ↓
Output Token IDs
        ↓
Tokenizer.decode()
        ↓
Human-readable text
```


````python
from google.colab import userdata
from huggingface_hub import login
from transformers import AutoTokenizer
````

````python
hf_token = userdata.get('HF_TOKEN')
login(hf_token, add_to_git_credential=True)
#add_to_git_credential=True means "also save my Hugging Face token into Git's credential storage
#so Git operations against Hugging Face repositories can authenticate automatically
````

````python
# Models
LLAMA = "meta-llama/Meta-Llama-3.1-8B-Instruct"   # Meta
PHI35 = "microsoft/Phi-3.5-mini-instruct"         # Microsoft
QWEN2 = "Qwen/Qwen2-7B-Instruct"                  # Alibaba
````

````python
tokenizer = AutoTokenizer.from_pretrained(LLAMA, trust_remote_code=True)
````

````python
text = "I am excited to to learn Generative AI . It's super cool !"
````

````python
# lets count the number of words in the sentence being tokenized
words = text.split() # this creates a list of words
print(f"Words = {words}")
print(f"Number of words = {len(words)}") # count number of words
````

````python
tokens = tokenizer.encode(text) # encode
````

````python
print(f"Tokens = {tokens}")
print(f"Number of tokens = {len(tokens)}") # count number of tokens
````

# Decode and Special Tokens

1. tokenizer.decode(tokens) is a method from Hugging Face's AutoTokenizer that converts token IDs (or tokens) back into a human-readable string—basically reversing the tokenization process.
2. The decoded string does not exactly match the original, since special token was added - ***<|begin_of_text|>***

***Note : These special tokens were used in training the model and hence it's different for each model.***

````python
tokenizer.decode(tokens) #decode
````

````python
#we can skip special tokens
tokenizer.decode(tokens, skip_special_tokens=True)
````

````python
# Batch token
text1 = "How are you ?"
text2 = "I am fine, thank you"
token1 = tokenizer.encode(text1) # encode
token2 = tokenizer.encode(text2) # encode
print(f"{text1} => after encoding => {token1}")
print(f"{text2} => after encoding => {token2}")
print(f"After batch decode = {tokenizer.batch_decode([token1, token2])}") # decode batch
````

````python
tokenizer.get_added_vocab() # these are the special tokens added by this LLM
# e.g., <|begin_of_text|> indicates the beginning of a prompt
````

````python
#This returns a dictionary mapping tokens to their IDs from
#the tokenizer's original vocabulary—before any additional tokens were added.

tokenizer.vocab
````

# Instruct variants of models

Many models have a variant that has been trained for use in Chats.  
These are typically labelled with the word "Instruct" at the end.  
They have been trained to expect prompts with a particular format that includes system, user and assistant prompts.  

There is a utility method `apply_chat_template` that will convert from the messages list format we are familiar with, into the right input prompt for this model.

````python
tokenizer = AutoTokenizer.from_pretrained(LLAMA, trust_remote_code=True)
````

````python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Tell a light-hearted joke for a room of Data Scientists"}
  ]

prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
print(f"The prompt is : {prompt}")
````

# Let us now try a few more models and compare


````python
phi3_tokenizer = AutoTokenizer.from_pretrained(PHI35)

tokens = phi3_tokenizer.encode(text)
print(f"Tokens : {tokens}")
print(100*"-")
decode = phi3_tokenizer.decode(tokens)
print(f"Decoded : {decode}")
print(100*"-")
batch_decode = phi3_tokenizer.batch_decode(tokens)
print(f"Batch Decode : {batch_decode}")
print(100*"-")

prompt = phi3_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
print(f"The prompt is : {prompt}") # simpler than LLAMA !
print(100*"-")
print(f"Number of tokens = {len(tokens)}") # count number of tokens
````

````python
qwen2_tokenizer = AutoTokenizer.from_pretrained(QWEN2)

tokens = qwen2_tokenizer.encode(text)
print(f"Tokens : {tokens}")
print(100*"-")
decode = qwen2_tokenizer.decode(tokens)
print(f"Decoded : {decode}")
print(100*"-")
batch_decode = qwen2_tokenizer.batch_decode(tokens)
print(f"Batch Decode : {batch_decode}")
print(100*"-")

prompt = qwen2_tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
print(f"The prompt is : {prompt}") # simpler than LLAMA !

print(100*"-")
print(f"Number of tokens = {len(tokens)}") # count number of tokens

````
